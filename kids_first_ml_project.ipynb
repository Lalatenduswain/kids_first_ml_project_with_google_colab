```python
# 🔤 FONT SETUP — Odia + Emoji support in charts
# ================================================
# (Run this cell first — only needed once per Colab session)

import subprocess, matplotlib, matplotlib.font_manager as fm

# Install Noto fonts (supports Odia script + emoji)
subprocess.run(['apt-get', 'install', '-y', 'fonts-noto', 'fonts-noto-color-emoji'],
               capture_output=True)

# Rebuild matplotlib font cache
fm._load_fontmanager(try_read_cache=False)

# Use Noto Sans (supports Odia) + fallback to DejaVu
matplotlib.rcParams['font.family'] = ['Noto Sans', 'Noto Sans Oriya', 'DejaVu Sans']

print("✅ Fonts installed! Odia text will now render correctly in charts.")
```

```python
# 🛒 SHOPPING LIST (tools ଆଣ!)
# ===============================

import numpy as np                              # ଗଣିତ ର ଦୋସ୍ତ 🔢
import matplotlib.pyplot as plt                # ଚିତ୍ର ଆଁକିବା ଲୋକ 🎨
from sklearn.tree import DecisionTreeClassifier # ← ଆମ AI ର brain! 🧠
from sklearn.metrics import accuracy_score      # ← marks ଦେଉଥିବା ଶିକ୍ଷକ 📝

print("=" * 40)
print("🤖 Robot: 'ସବୁ tools ଆସିଗଲା!'")
print("🎒 Backpack ଭର୍ତ୍ତି! Mission start! 🚀")
print("=" * 40)
print()
print("✅ numpy    — calculator ଭଳି (ଗଣିତ ପାଇଁ)")
print("✅ matplotlib — ଚିତ୍ରକର ଭଳି (chart ଆଁକିବ)")
print("✅ sklearn  — AI ର super brain! 🧠")
print()
print("🎉 READY TO ROLL!")
```

## Step 2 — 🥭🍊 ଫଳ ଡାଟା ତିଆର (Collect Evidence!)

> 🕵️ **Detective Robot:** "ଫଳ ଚିହ୍ନଟ ପାଇଁ ଆଗେ ଦେଖ — ଏହା ଆମ੍ବ ନା କମଳା?"
>
> **Clue 1 🔍:** ଆମ੍ବ ହାଲୁକା (80–110g), କମଳା ଭାରୀ (130–200g)!
> 
> **Clue 2 🎨:** ଆମ੍ବ ହଳଦିଆ (color=0), କମଳା... ଆଣ୍ଟା (color=1)!

| ଫଳ | ଓଜନ (ଗ୍ରାମ) | ରଙ୍ଗ code | ସନ୍ଦେହ ?
|---|---|---|---|
| 🍊 କମଳା | ଭାରୀ (130–200g) | 1 | "ମୁଁ ଗୋଲ ଓ ଭାରୀ! 😤" |
| 🥭 ଆମ୍ବ | ହାଲୁକା (80–110g) | 0 | "ମୁଁ slim ଓ ହଳଦିଆ! 😎" |

```python
# 🗂️ CASE FILE — Fruit Evidence Database
# ========================================
# format: [ଓଜନ(gram), ରଙ୍ଗ(0=ହଳଦିଆ, 1=କମଳା)]
#
# 🤖 Robot whispers: "ଏଇଟି AI ର question paper ভਲি!"

fruits_data = [
    # ଓଜନ,  ରଙ୍ଗ
    [150,    1],   # 🍊 କମଳା 1  — "ମୁଁ ଭାରୀ ଓ ଗୋଲ!"
    [170,    1],   # 🍊 କମଳା 2  — "ମୁଁ ଆହୁରି ଭାରୀ! 💪"
    [140,    1],   # 🍊 କମଳା 3  — "ଆଜି ଟିକେ diet ଅଛି"
    [160,    1],   # 🍊 କମଳା 4  — "Average ଭଳି ଅଛି"
    [180,    1],   # 🍊 କମଳା 5  — "Gym ଯାଏ 💪💪"
    [200,    1],   # 🍊 କମଳା 6  — "BOSS of oranges! 👑"
    [130,    1],   # 🍊 କମଳା 7  — "Smallest orange, still proud"
    [ 90,    0],   # 🥭 ଆମ୍ବ 1  — "ଆମ୍ବ ରାଜା! 👑"
    [100,    0],   # 🥭 ଆମ୍ବ 2  — "Alphonso ବଂଶ 😌"
    [ 80,    0],   # 🥭 ଆମ୍ବ 3  — "Lightweight champion! 🥊"
    [110,    0],   # 🥭 ଆମ୍ବ 4  — "ସବୁଠୁ ଭାରୀ ଆମ୍ବ 😤"
    [ 95,    0],   # 🥭 ଆମ୍ବ 5  — "ଠିକ ଠିକ ଅଛି"
    [105,    0],   # 🥭 ଆମ୍ବ 6  — "Round ଓ yummy! 😋"
    [ 85,    0],   # 🥭 ଆମ୍ବ 7  — "Smallest mango, big taste!"
]

# ✅ ଉତ୍ତର list — 0=ଆମ୍ବ, 1=କମଳା
labels = [1, 1, 1, 1, 1, 1, 1,   # ← ସବୁ 7 ଟି କମଳା (Team Orange 🍊)
          0, 0, 0, 0, 0, 0, 0]   # ← ସବୁ 7 ଟି ଆମ୍ବ  (Team Mango 🥭)

X = np.array(fruits_data)
y = np.array(labels)

print("🗂️  EVIDENCE COLLECTED!")
print("=" * 38)
print(f"🥭 Team Mango  (ଆମ୍ବ team) : {sum(y==0)} ଟି")
print(f"🍊 Team Orange (କମଳା team): {sum(y==1)} ଟି")
print(f"📦 ମୋଟ suspects           : {len(y)} ଟି")
print("=" * 38)
print("🕵️  Detective Robot: 'ଚାଲ! Case solve କରିବା!'")
```

## Step 3 — 📊 ସୁସ୍ପେକ୍ଟ Chart (Suspect Lineup!)

> 🕵️ **Detective Robot:** "ସବୁ ଫଳ ଗୁଡ଼ିକୁ line ରେ ଠିଆ କର! ଦେଖିବୁ — ଆମ੍ବ ହାଲୁକା, କମଳା ଭାରୀ!"
>
> *(ଯଦି ଫଳ ଚାଟ ଦେଖି ଠିକ ଚିହ୍ନଟ ହୁଏ, AI ସଫଳ! 🎯)*

```python
# 🎭 SUSPECT LINEUP CHART!
# ========================

plt.figure(figsize=(9, 5))

mango_weights  = [X[i][0] for i in range(len(y)) if y[i] == 0]
orange_weights = [X[i][0] for i in range(len(y)) if y[i] == 1]

# ଆମ੍ବ — ହଳଦିଆ dots
plt.scatter(range(len(mango_weights)), mango_weights,
            color='gold', s=300, label='🥭 ଆମ୍ବ (Slim guys!)', zorder=3,
            edgecolors='darkorange', linewidths=2)

# କମଳା — ଲାଲ dots
plt.scatter(range(len(orange_weights)), orange_weights,
            color='darkorange', s=300, label='🍊 କମଳା (Heavy squad!)', zorder=3,
            edgecolors='red', linewidths=2)

# ଦୁଇ team ମଝିରେ line
plt.axhline(y=120, color='red', linestyle='--', alpha=0.5, linewidth=2,
            label='🚔 Boundary (120g)')

plt.title('🥭 ଆମ୍ବ vs 🍊 କମଳା — Weight Showdown! 💪', fontsize=15, pad=12)
plt.ylabel('ଓଜନ (ଗ୍ରାମ) ↑ ଭାରୀ',  fontsize=12)
plt.xlabel('ଫଳ ସଂଖ୍ୟା (Fruit No.)', fontsize=12)
plt.legend(fontsize=11)
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()

print("💡 ଦେଖ! ଦୁଇ team ସ୍ପଷ୍ଟ ଅଲଗା!")
print("🤖 Robot: 'ଏଇଟି ଦେଖି ମୋ brain ଖୁସି ହୋଇଗଲା! 🥳'")
```

## Step 4 — 🧠 AI Brain ON! (Training Time!)

> 🤖 **Robot:** "ଏବେ ମୁଁ ଶିଖିବି! ଠିକ ଯେମିତି:"
>
> - 🍜 তুমি ଖାଇ ଖାଇ ଶିଖ (experience)
> - 📚 ମୁଁ data ଦେଖି ଶିଖ (training!)
>
> **🎓 School:** Data feeding into AI brain = 0.001 seconds!
> *(ତୁ 10 ବର୍ଷ ଧରି ଶିଖୁଛୁ... ମୁଁ ଏକ flash ରେ! 😏)*
>
> *(ହଁ ହଁ, exams ରେ ଚିଟ ନ ଦେ। AI ଶୀଖ!)*

```python
# 🧠 BRAIN ACTIVATION SEQUENCE
# =============================
import time

print("🔌 AI brain ଚାଲୁ ହେଉଛି...")
print("⚡ Loading... [██████████] 100% — DONE! (actually instant 😅)")
print()

# Decision Tree — ଭଲ detective ଭଳି ଗଛ ଭଳି ଚିନ୍ତା କରେ! 🌳
model = DecisionTreeClassifier(random_state=42)

# 🎓 TRAINING — AI ଡାଟା ଦେଖି ଶିଖୁଛି!
model.fit(X, y)

# AI ର self-test
predictions = model.predict(X)
accuracy    = accuracy_score(y, predictions) * 100

print("🎓 TRAINING REPORT CARD")
print("=" * 35)
print(f"📝 AI ର Marks   : {accuracy:.0f}% 📊")
print(f"📚 Questions    : {len(y)} ଟି ଫଳ")
print(f"✅ Correct      : {int(accuracy/100*len(y))} ଟି")
print("=" * 35)

if accuracy == 100:
    print()
    print("🎉 ୧୦୦%!! AI ଟି TOPPER!!")
    print("🤖 Robot: 'ଶିକ୍ଷକ ମୋତେ Gold Medal ଦିଅ! 🥇'")
    print("🏫 School: 'ତୁ Robot... Medal ନ ମିଳିବ 😤'")
    print("🤖 Robot: '...okay fine 😞'")
else:
    print(f"\n📖 AI ଆହୁରି ଶିଖୁଛି... {accuracy:.0f}% ହୋଇଗଲା!")
    print("🤖 Robot: 'ଆଉ data ଦିଅ! ଆଉ ଦିଅ!'")
```

## Step 5 — 🔍 Mystery Fruit Test! (Can AI Guess?)

> 🕵️ **Detective Robot:** "ଏବେ ଆସ — mystery fruits ଆସୁଛନ୍ତି! 🎭"
>
> "ସେମାନେ mask ପିନ୍ଧିଛନ୍ତି... କିନ୍ତୁ ଓଜନ ଲୁଚାଇ ପାରିବ ନାହାଁନ୍ତି! 😏"
>
> **ତୁ ନିଜ ଫଳ test କର!**
> ✏️ `new_fruits` list ରେ ଓଜନ ଓ ରଙ୍ଗ ବଦଳାଇ ଦେଖ!

```python
# 🔍 MYSTERY FRUIT INVESTIGATION!
# ================================
# ✏️ ଏଠି ନିଜ ଫଳ ଭରି ଦିଅ!
#    ଓଜନ: gram ରେ
#    ରଙ୍ଗ: 0=ହଳଦିଆ (ଆମ୍ବ ଭଳି), 1=କମଳା ରଙ୍ଗ

new_fruits = [
    [155, 1],   # 🎭 Mystery Fruit A — "ମୁଁ ଭାରୀ ଓ ଲାଲ!"
    [ 92, 0],   # 🎭 Mystery Fruit B — "ମୁଁ ହାଲୁକା ଓ ହଳଦିଆ!"
    [175, 1],   # 🎭 Mystery Fruit C — "ମୁଁ gym ଯाई! ଅnk দিna!"
    [ 88, 0],   # 🎭 Mystery Fruit D — "ଚୁପ ଚୁପ... (hiding)"
]

fruit_names  = ['A', 'B', 'C', 'D']
fruit_emojis = ['🎭', '🥸', '🤔', '😶']
label_map    = {0: '🥭 ଆମ୍ବ (MANGO!)',
                1: '🍊 କମଳା (ORANGE!)'}

results = model.predict(np.array(new_fruits))

print("🔍 DETECTIVE REPORT — MYSTERY FRUITS EXPOSED! 🎉")
print("=" * 48)
print(f"{'Fruit':<8} {'ଓଜନ':>6} {'ରଙ୍ଗ':>5}  →  ପରିଚୟ")
print("-" * 48)

for i, (fruit, result) in enumerate(zip(new_fruits, results)):
    color_name = 'ହଳଦିଆ' if fruit[1] == 0 else 'କମଳା'
    print(f"{fruit_emojis[i]} Fruit {fruit_names[i]}:  {fruit[0]}g  {color_name:>6}  →  {label_map[result]}")

print("=" * 48)
print()
print("🤖 Robot: 'ମୁଁ ସବୁ ଚିହ୍ନଟ କଲି! ଏବେ ଆଇସ୍ ଖିଆ ମିଳିବ? 🍦'")
print("👨‍🔬 Scientist: 'ତୁ ଯନ୍ତ୍ର... ଆଇସ୍ ଖୁଆ ମୁ! 😂'")
```
